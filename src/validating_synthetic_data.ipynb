This notebook is used to validate the distributions of the synthetic data

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances
import os
```

Cooling System

```python
#Reading the datasets from normal buses and finding the best comparison
sets = []
for i in range(31):
    sets.append(pd.read_csv(os.path.join('..\\Final_cooling_sys_data', 'final_cooling_dset_' + str(i) + '.csv')))
    sets[i].drop('Unnamed: 0', axis=1, inplace=True)
```

```python
original = pd.read_csv('..\\cooling_system.csv')
original.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
a = original.describe()
og_temp = a['65262-110'].to_numpy().reshape(1,-1)
og_fan = a['64817-1598'].to_numpy().reshape(1,-1)
og_lev = a['65263-111'].to_numpy().reshape(1,-1)
```

```python
#Comparing the distributions 
```

```python
scores = []
for i in range(31):
    curr = 0
    syn = sets[i]
    desc = syn.describe()
    fan = desc['64817-1598'].to_numpy().reshape(1,-1)
    temp = desc['65262-110'].to_numpy().reshape(1,-1)
    lev = desc['65263-111'].to_numpy().reshape(1,-1)
    curr += euclidean_distances(fan, og_fan)
    curr += euclidean_distances(temp, og_temp)
    curr += euclidean_distances(lev, og_lev)
    scores.append(curr)
min_val = min(scores)
idx = scores.index(min_val)
```
